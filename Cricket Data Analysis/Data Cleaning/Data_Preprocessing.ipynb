```python
#import necessary libraries
import pandas as pd
import json
import re
```

```python
with open('T20_Match_Results.json') as f:
    data = json.load(f) 
df_match = pd.DataFrame(data['matchSummary'])
df_match.head()
```

```python
len(df_match)
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match = df_match[df_match['winner']!= 'no result']
df_match.head()
```

```python
df_match.to_csv('Csv files/T20_Match_Results.csv', index = False)
```

**Batting Summary**

```python
with open('T20_Match_Batting_Summary.json') as f:
    data = json.load(f) 
    all_records = []
    for rec in data:
         all_records.extend(rec['battingSummary'])


df_batting = pd.DataFrame(all_records)
df_batting.head(50)
```

```python
df_batting.head(10)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "not_out" if x.strip().lower() == 'not out' else "out")
df_batting.head(50)
```

**Removing Special Characters from the table columns**

```python
# Define a lambda function to remove special characters using regular expressions
remove_special_chars = lambda x: re.sub(r'[^a-zA-Z0-9\s\(\)]', '', x)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(remove_special_chars)
df_batting.head(50)
```

```python
df_batting['dismissal'] = df_batting['dismissal'].apply(remove_special_chars)
df_batting.head(50)
```

```python
df_batting['match_id'] = df_batting['match_id'].str.replace('no.', '#')
df_batting.head(50)
```

```python
df_batting = df_batting[df_batting['match_name'].str.contains(r'vs\s+\S', case=False, na=False)]
df_batting.head(10)
```

```python
df_batting.to_csv('Csv files/T20_Match_Batting_Summary.csv', index = False)
```
